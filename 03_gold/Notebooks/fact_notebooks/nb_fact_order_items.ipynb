```python
%sql
    
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.fact_order_items AS
SELECT
    oi.order_id,
    oi.product_id,
    oi.seller_id,
    oi.price AS price,
    oi.freight_value AS freight_value,
    oi.price + oi.freight_value AS revenue
FROM ecommerce_orders.02_silver.order_items oi;
```
