```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.fact_orders AS
SELECT
    o.order_id,
    o.customer_id,
    CAST(o.order_purchase_timestamp AS DATE) AS order_date,
    o.delivery_days,
    o.is_late_delivery,
    o.order_status
FROM ecommerce_orders.02_silver.orders o;
```
