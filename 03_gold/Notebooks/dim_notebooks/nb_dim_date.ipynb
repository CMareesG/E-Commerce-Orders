```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.dim_date AS
SELECT DISTINCT
    CAST(order_purchase_timestamp AS DATE) AS date,
    YEAR(order_purchase_timestamp) AS year,
    MONTH(order_purchase_timestamp) AS month
FROM ecommerce_orders.02_silver.orders;
```
