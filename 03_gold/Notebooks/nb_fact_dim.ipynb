```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.dim_customer AS
SELECT DISTINCT
    customer_id,
    customer_unique_id,
    customer_zip_code_prefix
    FROM ecommerce_orders.02_silver.customers
```

```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.dim_products AS
SELECT DISTINCT
    product_id,
    product_category_name
FROM ecommerce_orders.02_silver.products;
```

```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.dim_sellers AS
SELECT DISTINCT
    seller_id
FROM ecommerce_orders.02_silver.sellers;
```

```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.geolocation AS
SELECT DISTINCT
    geolocation_zip_code_prefix,
    geolocation_city,
    geolocation_state
FROM ecommerce_orders.02_silver.geolocation;
```

```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.dim_date AS
SELECT DISTINCT
    CAST(order_purchase_timestamp AS DATE) AS date,
    YEAR(order_purchase_timestamp) AS year,
    MONTH(order_purchase_timestamp) AS month
FROM ecommerce_orders.02_silver.orders;
```

```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.fact_orders AS
SELECT
    o.order_id,
    o.customer_id,
    CAST(o.order_purchase_timestamp AS DATE) AS order_date,
    o.delivery_days,
    o.is_late_delivery,
    o.order_status
FROM ecommerce_orders.02_silver.orders o;
```

```python
%sql
    
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.fact_order_items AS
SELECT
    oi.order_id,
    oi.product_id,
    oi.seller_id,
    CAST(oi.price AS DECIMAL(10,2)) AS price,
    CAST(oi.freight_value AS DECIMAL(10,2)) AS freight_value,
    (CAST(oi.price AS DECIMAL(10,2)) + CAST(oi.freight_value AS DECIMAL(10,2))) AS revenue
FROM ecommerce_orders.02_silver.order_items oi;
```

```python
%sql
    
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.fact_payments AS
SELECT
    order_id,
    payment_types,
    payment_value
FROM ecommerce_orders.02_silver.order_payments;
```

```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.fact_reviews AS
SELECT
    order_id,
    review_score
FROM ecommerce_orders.02_silver.order_reviews;
```
