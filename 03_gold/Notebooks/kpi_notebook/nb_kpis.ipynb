1.Total revenue by month

```python
%sql
SELECT year, month, SUM(revenue) AS total_revenue
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY year, month
ORDER BY year, month;
```

2.Average Order Value (AOV)

```python
%sql
SELECT SUM(revenue)/COUNT(DISTINCT order_id) AS AOV
FROM ecommerce_orders.`03_gold`.curated_ecommerce;
```

3.Top 10 products by revenue


```python
%sql
SELECT product_category_name, SUM(revenue) AS revenue
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY product_category_name
ORDER BY revenue DESC
LIMIT 10;
```

4.Top 10 sellers by revenue

```python
%sql
SELECT seller_id,SUM(revenue) AS revenue
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY seller_id
ORDER BY revenue DESC
LIMIT 10;
    
```

5 Orders by customer city and state


```python
%sql
SELECT city,state, COUNT(DISTINCT order_id) AS total_orders
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY city,state;
```

6 New vs returning customers (monthly)


```python
%sql
WITH first_order AS (
    SELECT customer_id, MIN(order_date) AS first_date
    FROM ecommerce_orders.03_gold.curated_ecommerce
    GROUP BY customer_id
)

SELECT 
c.year,
c.month,
COUNT(DISTINCT CASE
WHEN c.order_date = f.first_date THEN c.customer_id END) AS new_customers,
COUNT(DISTINCT CASE
WHEN c.order_date > f.first_date THEN c.customer_id END) AS returning_customers
FROM ecommerce_orders.`03_gold`.curated_ecommerce c
JOIN first_order f
ON c.customer_id = f.customer_id
GROUP BY c.year, c.month;
```

7 Customer Lifetime Value (CLV)


```python
%sql
SELECT customer_id,SUM(revenue) AS CLV
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY customer_id;
```

8 Order fulfillment rate


```python
%sql
SELECT COUNT(CASE WHEN order_status = 'delivered' THEN 1 END)*100.0/COUNT(*) AS fulfillment_rate
FROM ecommerce_orders.`03_gold`.curated_ecommerce;
```

9 Late delivery percentage


```python
%sql
SELECT AVG(is_late_delivery) * 100 AS late_delivery_percentage
FROM ecommerce_orders.`03_gold`.curated_ecommerce;
```

10 Average delivery time (days)


```python
%sql
SELECT AVG(delivery_days) AS avg_delivery_days
FROM ecommerce_orders.`03_gold`.curated_ecommerce;
```

11 Payment method distribution


```python
%sql
SELECT payment_types, COUNT(*) as count
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY payment_types;
```

12 Average review score by product category


```python
%sql
SELECT product_category_name,AVG(review_score) AS avg_rating
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY product_category_name;
```

13 Average review score by seller


```python
%sql
SELECT seller_id, AVG(review_score) AS avg_rating
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY seller_id;
```

14 Percentage of low-rated orders (rating < 3)

```python
%sql
SELECT SUM(case WHEN review_score < 3 THEN 1 ELSE 0 END) * 100.0/COUNT(*) AS low_rating_percentage
FROM ecommerce_orders.`03_gold`.curated_ecommerce;
```
