```python
%sql
SELECT year, month, SUM(revenue) AS total_revenue
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY year, month
ORDER BY year, month;
```

```python
%sql
SELECT SUM(revenue)/COUNT(DISTINCT order_id) AS AOV
FROM ecommerce_orders.`03_gold`.curated_ecommerce;
```

```python
%sql
SELECT product_category_name, SUM(revenue) AS revenue
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY product_category_name
ORDER BY revenue DESC
LIMIT 10;
```

```python
%sql
SELECT seller_id,SUM(revenue) AS revenue
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY seller_id
ORDER BY revenue DESC
LIMIT 10;
    
```

```python
%sql
SELECT city,state, COUNT(DISTINCT order_id) AS total_orders
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY city,state;
```

```python
%sql
WITH first_order AS (
    SELECT customer_id, MIN(order_date) AS first_date
    FROM ecommerce_orders.03_gold.curated_ecommerce
    GROUP BY customer_id
)

SELECT 
c.year,
c.month,
COUNT(DISTINCT CASE
WHEN c.order_date = f.first_date THEN c.customer_id END) AS new_customers,
COUNT(DISTINCT CASE
WHEN c.order_date > f.first_date THEN c.customer_id END) AS returning_customers
FROM ecommerce_orders.`03_gold`.curated_ecommerce c
JOIN first_order f
ON c.customer_id = f.customer_id
GROUP BY c.year, c.month;
```
