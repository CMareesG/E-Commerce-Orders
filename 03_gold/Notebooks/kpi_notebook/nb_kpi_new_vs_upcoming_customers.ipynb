```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.kpi_new_vs_upcoming_customers AS
WITH first_order AS (
    SELECT customer_id, MIN(order_date) AS first_date
    FROM ecommerce_orders.03_gold.curated_ecommerce
    GROUP BY customer_id
)
SELECT 
    c.year,
    c.month,
    COUNT(DISTINCT CASE WHEN c.order_date = f.first_date THEN c.customer_id END) AS new_customers,
    COUNT(DISTINCT CASE WHEN c.order_date > f.first_date THEN c.customer_id END) AS returning_customers
FROM ecommerce_orders.03_gold.curated_ecommerce c
JOIN first_order f ON c.customer_id = f.customer_id
GROUP BY c.year, c.month;
```
