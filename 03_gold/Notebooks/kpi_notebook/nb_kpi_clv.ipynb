```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.03_gold.kpi_clv AS
SELECT
  customer_id,
  SUM(revenue) / COUNT(DISTINCT order_id) AS avg_order_value,
  COUNT(DISTINCT order_id) AS total_orders,
  COUNT(DISTINCT order_id) / COUNT(DISTINCT year) AS purchase_frequency,
  (SUM(revenue) / COUNT(DISTINCT order_id)) * (COUNT(DISTINCT order_id) / COUNT(DISTINCT year)) AS CLV
FROM ecommerce_orders.`03_gold`.curated_ecommerce
GROUP BY customer_id;
```
