```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.`03_gold`.curated_ecommerce AS 
SELECT 
  o.order_id,
  o.customer_id,
  o.order_date,
  d.year,
  d.month,
  c.customer_unique_id,
  g.geolocation_city as city,
  g.geolocation_state as state,
  p.product_category_name,

  oi.product_id,
  oi.price,
  oi.freight_value,
  (oi.price + oi.freight_value) as revenue,
  
  o.delivery_days,
  o.is_late_delivery,
  o.order_status,
  s.seller_id,

  pay.payment_types,
  pay.payment_value,
  
  r.review_score
FROM ecommerce_orders.03_gold.fact_orders o
JOIN ecommerce_orders.03_gold.dim_date d ON o.order_date = d.date
JOIN ecommerce_orders.03_gold.dim_customer c ON o.customer_id = c.customer_id
LEFT JOIN ecommerce_orders.03_gold.geolocation g on c.customer_zip_code_prefix = g.geolocation_zip_code_prefix 
JOIN ecommerce_orders.03_gold.fact_order_items oi ON o.order_id = oi.order_id
JOIN ecommerce_orders.03_gold.dim_products p ON oi.product_id = p.product_id
JOIN ecommerce_orders.03_gold.dim_sellers s ON oi.seller_id = s.seller_id
LEFT JOIN ecommerce_orders.03_gold.fact_payments pay ON pay.order_id=o.order_id
LEFT JOIN ecommerce_orders.03_gold.fact_reviews r ON r.order_id=o.order_id;
    
```
