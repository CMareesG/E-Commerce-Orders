```python
%sql
CREATE OR REPLACE TABLE ecommerce_orders.`02_silver`.geolocation AS
SELECT 
  CASE WHEN geolocation_zip_code_prefix IS NULL THEN 'Unknown' ELSE geolocation_zip_code_prefix END AS geolocation_zip_code_prefix,
  CASE WHEN geolocation_lat IS NULL THEN NULL ELSE CAST(geolocation_lat AS DOUBLE) END AS geolocation_lat,
  CASE WHEN geolocation_lng IS NULL THEN NULL ELSE CAST(geolocation_lng AS DOUBLE) END AS geolocation_lng,
  CASE WHEN geolocation_city IS NULL THEN 'Unknown' ELSE geolocation_city END AS geolocation_city,
  CASE WHEN geolocation_state IS NULL THEN 'Unknown' ELSE geolocation_state END AS geolocation_state
FROM ecommerce_orders.`01_bronze`.geolocation
```
