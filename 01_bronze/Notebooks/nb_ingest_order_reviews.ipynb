```python
import pandas as pd
import io

s3_path = "s3://ecommerce-order-s3-bucket-781856668143-ap-southeast-1-an/OlistDataset/olist_order_reviews_dataset.csv.xlsx"

file_content = spark.read.format("binaryFile").load(s3_path).collect()[0]["content"]

pandas_df = pd.read_excel(io.BytesIO(file_content))

for col in pandas_df.columns:
    if pandas_df[col].dtype == 'object':
        pandas_df[col] = pandas_df[col].astype(str)

order_items_df = spark.createDataFrame(pandas_df)

order_items_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("ecommerce_orders.01_bronze.order_reviews")
```
